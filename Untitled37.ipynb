```python
df_filled = df.fillna(0)
print("Dataframe with missing values replace by 0",df_filled)
```

```python
df['date'] = pd.to_datetime(df['date'])
print(df)
```

```python
df['total_deaths'] = pd.to_numeric(df['total_deaths'],errors='coerce')
df['total_cases'] = pd.to_numeric(df['total_cases'],errors='coerce')
df['total_deaths_to_total_cases_Ratio']=df['total_deaths']/df['total_cases']
print(df)
```

```python
import matplotlib.pyplot as plt

df['gdp_per_capita'] = pd.to_numeric(df['gdp_per_capita'])
plt.hist(df['gdp_per_capita'],color="blue")
plt.title("Distribution of gdp per capita")
plt.xlabel("gdp per capita")
plt.ylabel("frequaecy")
plt.show
```

```python
df['total_cases'] = pd.to_numeric(df['total_cases'])
df['gdp_per_capita'] = pd.to_numeric(df['gdp_per_capita'])
plt.scatter(df['total_cases'],df['gdp_per_capita'],color="orange")
plt.title("Scatter plot total_cases vs gdp_per_capita")
plt.xlabel("total cases")
plt.ylabel("gpd per capita")
plt.show
```

```python
df['total_cases'] = pd.to_numeric(df['total_cases'])
df_continent_total_cases = df.groupby('continent')['total_cases'].sum().reset_index()
plt.figure(figsize=(10,6))
plt.bar(df_continent_total_cases['continent'],df_continent_total_cases['total_cases'],color='skyblue')
plt.title('Total cases by continent')
plt.xlabel('continent')
plt.ylabel('total cases')
plt.show
```
